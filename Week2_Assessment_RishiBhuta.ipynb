 Q1. What's the difference between a Series and a DataFrame? Give one operation that works on both, and one that works only on a DataFrame.



Q1Ans: A Series is a one-dimensional labeled array that stores a single column of data. A DataFrame is a two-dimensional table made up of multiple Series sharing the same index.
Both support operations like .head() to view the first few records. An operation like .merge() works only on DataFrames because it combines two tables.

 Q2. What does `inplace=True` do? Why is the pandas team discouraging its use in newer versions? Give one example where you'd avoid it.


Q2Ans: inplace=True modifies the original DataFrame instead of returning a new one. Newer versions of pandas discourage its use because it can make code harder to read and does not always improve performance.

I would avoid it when cleaning data so I can keep the original DataFrame unchanged for comparison or debugging.

 Q3. Explain the difference between `.fillna()` and `.dropna()`. When would a data analyst pick one over the other? Be specific — give a concrete situation for each.

Q3Ans: fillna() replaces missing values with a chosen value, while dropna() removes rows or columns containing missing values.

A data analyst would use fillna() when missing values can be reasonably estimated, such as replacing missing salaries with the average salary. dropna() is useful when only a few incomplete rows exist and removing them will not affect the analysis.

 Q4. What does `pd.to_numeric(col, errors='coerce')` do that plain `pd.to_numeric(col)` does NOT? Which one would you use on real-world messy data, and why?


Q4Ans: pd.to_numeric(col, errors='coerce') converts invalid values into NaN instead of raising an error. Without errors='coerce', pandas stops execution if it encounters non-numeric values.

For real-world datasets, I would use errors='coerce' because messy data often contains values like "NA" or "Unknown".


 Q5. Explain `.loc` vs `.iloc` with one example each. Then describe a case where `df.loc[0]` and `df.iloc[0]` return different rows.
 

Q5Ans: .loc selects rows using labels, while .iloc selects rows using integer positions.
Example:
df.loc[0]
returns the row with label 0.

df.iloc[0]
returns the first row regardless of its label.

They return different rows when the DataFrame index is not sequential. For example, if the index is [5, 6, 7], df.loc[0] will raise an error, while df.iloc[0] returns the first row.

 Q6. What is the difference between `.map()`, `.apply()`, and `.replace()` for a Series? When would you reach for each?



Q6Ans: map() is used to transform values using a dictionary, function, or another Series. apply() applies a custom function to every value in the Series. replace() substitutes specific values with new values.

I would use:

map() for mapping categories like city aliases.
apply() for custom calculations.
replace() for replacing fixed values such as "NA" with NaN.

 Q7. What is the `SettingWithCopyWarning`? Give one example of code that triggers it and one way to fix it. (Hint: the chained-assignment pattern.)

Q7Ans: *AI-Assisted: Explanation reviewed and refined with ChatGPT.*

 SettingWithCopyWarning occurs when modifying a filtered DataFrame that may be a copy instead of the original DataFrame. This can lead to unexpected results.

Example that triggers the warning:
filtered = df[df["rating"] > 4]
filtered["rating"] = filtered["rating"] * 2
One way to fix it is by creating an explicit copy:

filtered = df[df["rating"] > 4].copy()
filtered["rating"] = filtered["rating"] * 2
Another solution is to modify the original DataFrame using .loc.

Q8. Explain `pd.cut()` vs `pd.qcut()`. If you have annual income data and want to label customers as Low / Mid / High earners, which one fits better? Why?


Q8Ans: *AI-Assisted: Explanation reviewed and refined with ChatGPT.*

pd.cut() divides data into intervals with fixed value ranges, while pd.qcut() divides data into groups containing approximately equal numbers of observations.

For annual income labels such as Low, Mid, and High earners, I would use pd.qcut() because it creates balanced groups even when the income distribution is uneven.

## Cleaning Pipeline (do this ONCE)

```python
# Cleaning Dataset

import pandas as pd
import numpy as np

df = pd.read_csv("indian_startups_fundingDADS.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Clean startup and city names
df["startup_name"] = df["startup_name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()

# City aliases
city_map = {
    "Bangalore": "Bengaluru",
    "Blr": "Bengaluru",
    "Bombay": "Mumbai",
    "Gurgaon": "Gurugram",
    "Delhi": "New Delhi",
    "Calcutta": "Kolkata",
    "Madras": "Chennai"
}

df["city"] = df["city"].replace(city_map)

# Industry aliases
industry_map = {
    "Fin Tech": "Fintech",
    "FinTech": "Fintech",
    "Ed Tech": "Edtech",
    "Health Tech": "HealthTech",
    "Healthtech": "HealthTech",
    "E-Commerce": "Ecommerce",
    "Saas": "SaaS",
    "Ai": "AI"
}

df["industry"] = df["industry"].str.strip().replace(industry_map)

# Funding Stage aliases
stage_map = {
    "Seed Round": "Seed",
    "Seed": "Seed",
    "Pre-Seed": "Pre-Seed",
    "Series A": "Series A",
    "Series B": "Series B",
    "Series C": "Series C"
}

df["funding_stage"] = df["funding_stage"].str.strip().replace(stage_map)

# Clean Amount
df["amount"] = (
    df["amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("INR", "", regex=False)
    .str.replace("Cr", "", regex=False)
    .str.replace("Crore", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
    .replace("Undisclosed", np.nan)
)

df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

# Date

df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce",
    dayfirst=True
)

# Remove duplicates
df = df.drop_duplicates()
```

### Q29. Add `funding_efficiency = amount / (2024 - founded_year)`. Sort descending. Print top 10: name, industry, amount, founded_year, funding_efficiency.

### Q30. Using `.apply()` + lambda, add `funding_tier`:
  - >=1000 Cr -> 'Unicorn'
  - >=500 Cr -> 'Mega'
  - >=100 Cr -> 'Large'
  - >=10 Cr -> 'Mid'
  - else (incl. NaN) -> 'Small / Unknown'
Print count per tier.

```python
# Answer Q30

df["funding_tier"] = df["amount"].apply(
    lambda x:
        "Unicorn" if x >= 1000 else
        "Mega" if x >= 500 else
        "Large" if x >= 100 else
        "Mid" if x >= 10 else
        "Small / Unknown"
)

print(df["funding_tier"].value_counts())
```

### Q32. `investors` column has multiple investors separated by ', '. Add `investor_count` via `.apply()` + lambda (NaN -> 0). Print `value_counts()`.

## Aggregation & Grouping (Q33–Q37)

### Q33. Average funding per industry (ignore NaN), sorted descending. Print top 5.

```python
# Answer Q33

result = (
    df.groupby("industry")["amount"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

print(result)
```

### Q34. Per city, single `.agg()` dict for: (a) startup count (b) total funding (c) avg funding per startup. Print top 10 cities by count.

```python
# Answer Q34

result = (
    df.groupby("city")
      .agg({
          "startup_name": "count",
          "amount": ["sum", "mean"]
      })
)

result.columns = [
    "Number of Startups",
    "Total Funding",
    "Average Funding"
]

result = result.sort_values(
    by="Number of Startups",
    ascending=False
)

print(result.head(10))
```

### Q35. Industry with highest average funding, but only industries with >=15 startups. Print industry, count, avg funding.

```python
# AI-Assisted: Grouping logic reviewed with ChatGPT.

industry_stats = (
    df.groupby("industry")
      .agg(
          count=("startup_name", "count"),
          average_funding=("amount", "mean")
      )
)

industry_stats = industry_stats[
    industry_stats["count"] >= 15
]

result = industry_stats.sort_values(
    by="average_funding",
    ascending=False
)

print(result.head(1))
```

### Q36. `pd.crosstab()`: funding_tier distribution across top 5 industries by count. Rows=industry, cols=funding_tier, values=count.

```python
# Answer Q36

top5 = (
    df["industry"]
      .value_counts()
      .head(5)
      .index
)

result = pd.crosstab(
    df[df["industry"].isin(top5)]["industry"],
    df[df["industry"].isin(top5)]["funding_tier"]
)

print(result)
```

### Q37. Per industry, the highest-funded startup (groupby + `idxmax()` on amount). Print startup_name, industry, city, amount.

```python
# AI-Assisted: idxmax() solution reviewed with ChatGPT.

idx = df.groupby("industry")["amount"].idxmax()

result = df.loc[
    idx,
    [
        "startup_name",
        "industry",
        "city",
        "amount"
    ]
]

print(result)
```

## Real Industry Problems (Q38–Q40)

### Q38. 'Emerging player' = founded >=2019, amount >=100 Cr, industry in [Fintech, Edtech, HealthTech]. Sort by amount descending. Print name, industry, city, founded_year, amount. Add a 1-line note on why a VC analyst would care.

```python
# Answer Q38

result = df[
    (df["founded_year"] >= 2019) &
    (df["amount"] >= 100) &
    (
        df["industry"].isin(
            [
                "Fintech",
                "Edtech",
                "HealthTech"
            ]
        )
    )
]

result = result.sort_values(
    by="amount",
    ascending=False
)

print(
    result[
        [
            "startup_name",
            "industry",
            "city",
            "founded_year",
            "amount"
        ]
    ]
)
```

### Q39. City with highest TOTAL funding, considering only cities with >=20 startups. Print city, count, total funding.

```python
# AI-Assisted: Aggregation logic reviewed with ChatGPT.

city_stats = (
    df.groupby("city")
      .agg(
          count=("startup_name", "count"),
          total_funding=("amount", "sum")
      )
)

city_stats = city_stats[
    city_stats["count"] >= 20
]

result = city_stats.sort_values(
    by="total_funding",
    ascending=False
)

print(result.head(1))
```

### Q40. For each (city, industry) combo, count startups. Print top 10 most-saturated combinations.

```python
# Answer Q40

result = (
    df.groupby(
        [
            "city",
            "industry"
        ]
    )
    .size()
    .reset_index(name="startup_count")
    .sort_values(
        by="startup_count",
        ascending=False
    )
)

print(result.head(10))
```
